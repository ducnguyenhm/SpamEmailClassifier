```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
```

```python
# Load training and testing data
train_df = pd.read_csv('./Data/spam_train.csv')  
test_df = pd.read_csv('./Data/spam_test.csv')   

# Inspect the data
train_df.head(5)
```

```python
X_train = train_df.drop('class', axis=1)
y_train = train_df['class']

X_test = test_df.drop('class', axis=1)
y_test = test_df['class']
```

```python
X_train
```

```python
y_train
```

```python
# --------------------------------------------------------------------------
# 1) K-Nearest Neighbors (KNN)
# --------------------------------------------------------------------------
# Define a range of K values to try
k_values = [1, 3, 5, 7, 9, 11, 13, 15]

knn_results = []
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=5, scoring='accuracy')
    knn_results.append((k, scores.mean()))

best_k, best_knn_cv_acc = max(knn_results, key=lambda x: x[1])
print("KNN cross-validation results:")
for k, acc in knn_results:
    print(f"  k={k}, CV Accuracy={acc:.4f}")
print(f"Best k: {best_k}, CV Accuracy: {best_knn_cv_acc:.4f}\n")

# Plot to compare the cross-validation accuracy of different K values
k_values, accuracies = zip(*knn_results)
plt.figure(figsize=(10, 6))
plt.plot(k_values, accuracies, marker='o', linestyle='-', color='b', label='KNN CV Accuracy')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Cross-Validation Accuracy')
plt.title('KNN 5-fold Cross-Validation Accuracy for Different k Values')
plt.legend()
plt.grid(True)
plt.show()

# Train the KNN model with the best K value on the entire training set
best_knn = KNeighborsClassifier(n_neighbors=best_k)
best_knn.fit(X_train, y_train)

y_test_pred_knn = best_knn.predict(X_test)
knn_test_acc = accuracy_score(y_test, y_test_pred_knn)
print(f"KNN Test Accuracy: {knn_test_acc:.4f}")
```

```python
# --------------------------------------------------------------------------
# 2) Logistic Regression
# --------------------------------------------------------------------------

# Define a range of C values to try (inverse of regularization strength)
c_values = [0.001, 0.01, 0.1, 1, 10, 100]

lr_results = []
for c in c_values:
    lr = LogisticRegression(max_iter=1000, C=c, solver='liblinear')
    scores = cross_val_score(lr, X_train, y_train, cv=5, scoring='accuracy')
    lr_results.append((c, scores.mean()))

best_c_lr, best_lr_cv_acc = max(lr_results, key=lambda x: x[1])
print("Logistic Regression cross-validation results:")
for c, acc in lr_results:
    print(f"  C={c}, CV Accuracy={acc:.4f}")
print(f"Best C: {best_c_lr}, CV Accuracy: {best_lr_cv_acc:.4f}\n")

# Plot to compare the cross-validation accuracy of different C values for Logistic Regression
c_values, accuracies = zip(*lr_results)
plt.figure(figsize=(10, 6))
plt.plot(c_values, accuracies, marker='o', linestyle='-', color='r', label='Logistic Regression CV Accuracy')
plt.xlabel('Inverse of Regularization Strength (C)')
plt.ylabel('Cross-Validation Accuracy')
plt.title('Logistic Regression 5-fold Cross-Validation Accuracy for Different C Values')
plt.xscale('log')
plt.legend()
plt.grid(True)
plt.show()

# Train the Logistic Regression model with the best C value on the entire training set
best_lr = LogisticRegression(C=c, solver='liblinear')
best_lr.fit(X_train, y_train)

y_test_pred_lr = best_lr.predict(X_test)
lr_test_acc = accuracy_score(y_test, y_test_pred_lr)
print(f"Logistic Regression Test Accuracy: {lr_test_acc:.4f}")
```
